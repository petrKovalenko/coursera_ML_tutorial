```python
import numpy as np
```

```python
matr = np.random.normal(loc=1,scale=10, size=(100,5))
print(matr)
```

```python
mean = np.mean(matr, axis=0)
print("mean:")
print(mean)
std = np.std(matr, axis=0)
print("std:")
print(std)
matr_norm = ((matr - mean) / std)
print("matr:")
print(matr_norm)
```

```python
sum_strings = np.sum(matr, axis=1)
print("sum_strings")
print(sum_strings)
print("sum more than 20:")
print(np.nonzero(sum_strings > 20))
```
