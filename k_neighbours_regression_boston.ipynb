```python
import numpy as np
import pandas as ps
import sklearn
import sklearn.model_selection
import sklearn.neighbors
print(sklearn.__version__)
```

```python
boston_ds = sklearn.datasets.load_boston()
```

```python
print(boston_ds.data)
boston_data_scaled = sklearn.preprocessing.scale(boston_ds.data)
print(boston_data_scaled)
```

```python
print(boston_ds.target)
```

```python
linespaces = np.linspace(1, 10, 200)
print(linespaces)
```

```python
cross_validator = sklearn.model_selection.KFold(n_splits=5, shuffle=True, random_state=42)
max_sq_error = -10000000000000
min_p = -1
for cur_p in linespaces:
    print(str(cur_p))
    cur_regressor = sklearn.neighbors.KNeighborsRegressor(n_neighbors=5, weights='distance', metric='minkowski', p=cur_p)
    cur_sq_error = sklearn.model_selection.cross_val_score(cur_regressor, boston_data_scaled, boston_ds.target, cv=cross_validator, scoring="neg_mean_squared_error")
    print(cur_sq_error)
    cur_avg_error = sum(cur_sq_error) / len(cur_sq_error)
    print(cur_avg_error)
    if cur_avg_error > max_sq_error:
        max_sq_error = cur_avg_error
        min_p = cur_p
print("Minimum error:", max_sq_error)
print("P of minimum error:", min_p)
```
