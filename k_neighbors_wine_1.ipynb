```python
import numpy as np
import pandas as ps
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
#wine_data = ps.read_csv("C:\\Users\\p.kovalenko\\Documents\\GitHub\\coursera_ML_tutorial\\coursera_ML_tutorial\\wine.data")
wine_data = ps.read_csv("C:\\Users\\asus\\Desktop\\coursera_ML_tutorial\\wine.data")
```

```python
print(wine_data)
```

```python
#my_cols = set(wine_data.columns)
#print(my_cols)
#matrix_cols = my_cols.remove("Class")
variable_matrix = wine_data.drop(["Class"], axis = 1)
aim_column = list(wine_data["Class"])
#print(variable_matrix)
print(aim_column)
#print(list(aim_column))
```

```python
#cross_validator = KFold(n_splits=5, shuffle=True)
cross_validator = KFold(n_splits=5, shuffle=True, random_state=42)
for train_index, test_index in cross_validator.split(variable_matrix):
    print("TRAIN:", train_index, " TEST:", test_index)
```

```python
k_classifier = KNeighborsClassifier(n_neighbors=7)
print(cross_val_score(k_classifier, variable_matrix, aim_column, cv=cross_validator, scoring="accuracy"))
```

```python
max_accuracy = 0
max_index = 0
for k in range(1,51):
    print(str(k))
    cur_classifier = KNeighborsClassifier(n_neighbors=k)
    cur_accuracy_array = cross_val_score(cur_classifier, variable_matrix, aim_column, cv=cross_validator, scoring="accuracy")
    print(cur_accuracy_array)
    cur_avg_acc = sum(cur_accuracy_array) / len(cur_accuracy_array)
    print(cur_avg_acc)
    if cur_avg_acc > max_accuracy:
        max_accuracy = cur_avg_acc
        max_index = k
print("Max accuracy:", max_accuracy)
print("Max index:", max_index)
```
